# **Feature selection**

## We need to encode data before feature selection

```python
df1=df.copy()
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for col in df1.columns:
  df1[col]=le.fit_transform(df1[col])
df1
```

## **CHI SQUARE TEST** is applied to the features with **categorical data**



```python
x_categorical=df1[['workclass','education','education.num','marital.status','occupation','relationship','race','sex','native.country']]
y1=df1['income']
y1
```

## Applying chi square test

**f_value:** Measures the strength of the relationship between variables

**p_value:** assesses the significance of that relationship

f_value should be higher and p_value should be lower

if f_value is lower and the p_value is higher, it suggests that there may not be a significant relationship between the variables being analyzed, and the observed differences may be due to chance or random variation.So the column has to be removed

```python
from sklearn.feature_selection import chi2
score1=chi2(x_categorical,y1)
score1
```

```python
f_value1=pd.Series(score1[0],index=x_categorical.columns)
f_value1.sort_values(ascending=False)
```

```python
p_value1=pd.Series(score1[1],index=x_categorical.columns)
p_value1.sort_values(ascending=False)
```

## Here we can say that the column **native.country** has the lowest f value and highest p value.Hence it needs to be dropped

## **ANNOVA TEST** is applied to the features with **continous numerical data**

```python
x_continuos=df1[['age','fnlwgt','capital.gain','capital.loss','hours.per.week']]
from sklearn.feature_selection import f_classif
score2=f_classif(x_continuos,y1)
score2
```

```python
f_value2=pd.Series(score2[0],index=x_continuos.columns)
f_value2.sort_values(ascending=False)
```

```python
p_value2=pd.Series(score2[1],index=x_continuos.columns)
p_value2.sort_values(ascending=False)
```

## Here we can say that the column **fnlwgt** has the lowest f value and highest p value.Hence it needs to be dropped

```python
abc=df[['education','education.num']]
for i in abc.columns:
  plt.figure(figsize=(15,5))
  sns.countplot(x=i,data=abc,order=abc[i].value_counts().index)
```

Here we can say that the column education and education.num implies the same

Hence we can drop **education.num**

## **Droping Columns**



```python
df.drop(['education','native.country','fnlwgt'],axis=1,inplace=True)
df
```

## **TO IDENTIFY THE COLUMNS WITH '?'**

```python
for column in df.columns:
    if '?' in df[column].values:
      print(column)
```

## **Converting '?' to missing values**

```python
df.loc[df.workclass=='?','workclass']=np.NaN
df.loc[df.occupation=='?','occupation']=np.NaN
```

## To balance it we use **Oversampling** technique

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE()
x1,y1=smote.fit_resample(x,y)
```

## **Splitting into training and testing data**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.30,random_state=42)
```

## **Normalization**

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```
